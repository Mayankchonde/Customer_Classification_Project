# Chi-square distribution
# By hypothesis testing
# Consider null and alternative hypothesis
# H0: The grouping variables have no association or correlation amongst them.
# H1: The variables are associated with each other and happen to have a correlation between the variables.

# By the p-value we interpret the result.If P-value is greater then level of significance which is 0.05, then we accept the null hypothesis and say the correlation of one independent variable is correlate to dependent variable.

```python
from scipy.stats import chi2_contingency 
import numpy as np
chsqt = pd.crosstab(ca['checking_status'], ca['class'], margins=True)
chsqt
value = np.array([chsqt.iloc[0][0:5].values,chsqt.iloc[1][0:5].values,chsqt.iloc[2][0:5].values,chsqt.iloc[3][0:5].values])
print(chi2_contingency(value)[0:3])
p=chi2_contingency(value)[1]
significance_level = 0.05
print("p value: " + str(p)) 
if p <= significance_level: 
    print('Reject NULL HYPOTHESIS') 
else: 
    print('Fail to Reject NULL HYPOTHESIS') 
```
